Se toman los datos del promedio móvil y el rendimiento para realizar la predicción del precio del maíz mediante simulaciones. 

```python
x = df_pe.iloc[:, 6].values
y = df_pe.iloc[:,7].values
```

```python
def fun_obj(B, x, y, n):
    return np.mean((y-np.array([x**i for i in range(n+1)]).T.dot(B))**2)
```

```python

B01 = np.zeros(2)
ajuste1 = opt.minimize(fun_obj, B01, args=(x,y,1))
B02 = np.zeros(3)
ajuste2 = opt.minimize(fun_obj, B02, args=(x,y,2))
B03 = np.zeros(4)
ajuste3 = opt.minimize(fun_obj, B03, args=(x,y,3))
B04 = np. zeros(5)
ajuste4 = opt.minimize(fun_obj,B04,args=(x,y,4))
B05 = np. zeros(6)
ajuste5 = opt.minimize(fun_obj,B05,args=(x,y,5))
B06 = np.zeros(7)
ajuste6=opt.minimize(fun_obj,B06,args=(x,y,6))
B20 =np.zeros (16)
ajuste20=opt.minimize(fun_obj,B20,args=(x,y,15))
```

```python

yhat1 = ajuste1.x[0]+ajuste1.x[1]*x
yhat2 = ajuste2.x[0]+ajuste2.x[1]*x+ajuste2.x[2]*x**2
yhat3 = ajuste3.x[0]+ajuste3.x[1]*x+ajuste3.x[2]*x**2+ajuste3.x[3]*x**3
yhat4= ajuste4.x[0]+ajuste4.x[1]*x+ajuste4.x[2]*x**2+ajuste4.x[3]*x**3+ajuste4.x[4]*x**4 
yhat5 = ajuste5.x[0]+ajuste5.x[1]*x+ajuste5.x[2]*x**2+ajuste5.x[3]*x**3+ajuste5.x[4]*x**4+ajuste5.x[5]*x**5
yhat6 = ajuste6.x[0]+ajuste6.x[1]*x+ajuste6.x[2]*x**2+ajuste6.x[3]*x**3+ajuste6.x[4]*x**4+ajuste6.x[5]*x**5+ajuste6.x[6]*x**6
yhat20= ajuste20.x[0]+ajuste20.x[1]*x+ajuste20.x[2]*x**2+ajuste20.x[3]*x**3+ajuste20.x[4]*x**4+ajuste20.x[5]*x**5+ajuste20.x[6]*x**6 +ajuste20.x[7]*x**7+ajuste20.x[8]*x**8+ajuste20.x[9]*x**9+ajuste20.x[10]*x**10+ajuste20.x[11]*x**11+ajuste20.x[12]*x**12+ajuste20.x[13]*x**13+ajuste20.x[14]*x**14+ajuste20.x[15]*x**15
plt.figure(figsize=(8,6))
plt.plot(x, y,"o" ,label = 'datos reales')
plt.plot(x,yhat20,label='ajuste6')
plt.xlabel('meses')-
plt.ylabel('precio de cierre')
plt.legend(loc='best')
plt.show()
```

```python
error1 = np.mean((y-yhat1)**2)
error2 = np.mean((y-yhat2)**2)
error3 = np.mean((y-yhat3)**2)
error4 = np.mean((y-yhat4)**2). zeros(6)
ajuste5 = opt.minimize(fun_obj,B05,args=(x,y,5))
B06 = np.zeros(7)
ajuste6=opt.minimize(fun_obj,B06,args=(x,y,6))
B20 =np.zeros (16)
ajuste20=opt.minimize(fun_obj,B20,args=(x,y,15))
error5 = np.mean((y-yhat5)**2)
error20 = np.mean((y-yhat20**2))

plt.figure(figsize=(8,6))
plt.plot([1,2,3,4,5,6],[error1,error2,error3,error4,error5,error20], 'o')
plt.xlabel('orden del polinomio')
plt.ylabel('error cuadrático')
plt.show()
```
